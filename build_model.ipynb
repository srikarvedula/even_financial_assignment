```python
import pandas as pd
import sqlite3
database_name = "even_financial.db"
# Read sqlite query results into a pandas DataFrame
con = sqlite3.connect(database_name)
leads_df = pd.read_sql_query("SELECT * from leads", con)
print(leads_df)
clicks_df = pd.read_sql_query("SELECT * from clicks", con)
print(clicks_df)
offers_df = pd.read_sql_query("SELECT * from offers", con)
print(offers_df)
```

```python
df=leads_df.merge(offers_df, on='lead_uuid', how='outer')
df
```

```python
new_df=df.merge(clicks_df,on='offer_id', how='outer')
new_df
```

```python
new_df['clicked_at'] = new_df['clicked_at'].fillna(0)
new_df
```

```python
def convert_datetime_to_clicked(x):
    if isinstance(x,str):
        return 1
    else:
        return x
    
```

```python
new_df['clicked_at']=[convert_datetime_to_clicked(x) for x in new_df['clicked_at']]
new_df
```

```python
new_df['clicked_at'].unique()
```

```python
import seaborn as sns
x = new_df['clicked_at'].value_counts()
print(x)
sns.barplot(x.index,x)
```

```python
new_df=new_df.drop(['lead_uuid','offer_id','lender_id'],axis=1)
new_df
```

```python
threshold= 9320
new_df_list=[]
df_list=list(new_df['clicked_at'].unique())
for i in df_list:
    x=new_df['clicked_at'].value_counts()[i]
    if x > threshold:
        y=new_df[new_df['clicked_at']==i].sample(n=threshold)
        new_df_list.append(y)
    else:
        y=new_df[new_df['clicked_at']==i]
        new_df_list.append(y)
        
final_df=pd.concat(new_df_list)
final_df
```

```python
import seaborn as sns
x = final_df['clicked_at'].value_counts()
print(x)
sns.barplot(x.index,x)
```

```python
final_df.to_csv('final_df.csv')
```

```python
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X = final_df.loc[:, final_df.columns != 'clicked_at']
y = final_df.loc[:, final_df.columns == 'clicked_at']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
logisticRegr = LogisticRegression(solver='liblinear',C=15)
```

```python
logisticRegr.fit(x_train, y_train)
```

```python
predictions = logisticRegr.predict(x_test)
```

```python
predictions
```

```python
score = logisticRegr.score(x_test, y_test)
print(score)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
```

```python
cm = metrics.confusion_matrix(y_test, predictions)
print(cm)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedStratifiedKFold
```

```python
model = LogisticRegression()
solvers = ['newton-cg', 'lbfgs', 'liblinear']
penalty = ['l2']
c_values = [100, 10, 1.0, 0.1, 0.01]
# define grid search
grid = dict(solver=solvers,penalty=penalty,C=c_values)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
grid_search = GridSearchCV(estimator=model, param_grid=grid, n_jobs=-1, cv=cv, scoring='accuracy',error_score=0)
grid_result = grid_search.fit(x_train, y_train)
# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```
